Dari setiap data yang sudah di `describe()` dapat dilihat bahwa kemungkinan data memiliki Outliers. Untuk memastikannya kita lakukan pengecheckan outliers menggunakan **metode IQR**

Check Outlier, dan menghitung outlier yang ada di setiap kolom (di setiap data kualitas udara)

```python
#Column yang akan di check
columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

jumlah_outliers = {}

#Menghitung outlier aotizhongxin_df
for column in columns:
    q25, q75 = np.percentile(aotizhongxin_df[column], 25), np.percentile(aotizhongxin_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in aotizhongxin_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'aotizhongxin_{column}'] = len(outliers)  # Simpan jumlah outlier

#Menghitung outlier changping_df
for column in columns:
    q25, q75 = np.percentile(changping_df[column], 25), np.percentile(changping_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in changping_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'changping_{column}'] = len(outliers)  # Simpan jumlah outlier

#Menghitung outlier dingling_df
for column in columns:
    q25, q75 = np.percentile(dingling_df[column], 25), np.percentile(dingling_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in dingling_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'dingling_{column}'] = len(outliers)  # Simpan jumlah outlier\

#Menghitung outlier dongsi_df
for column in columns:
    q25, q75 = np.percentile(dongsi_df[column], 25), np.percentile(dongsi_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in dongsi_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'dongsi_{column}'] = len(outliers)  # Simpan jumlah outlier

#Menghitung outlier guanyuan_df
for column in columns:
    q25, q75 = np.percentile(guanyuan_df[column], 25), np.percentile(guanyuan_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in guanyuan_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'guanyuan_{column}'] = len(outliers)  # Simpan jumlah outlier

#Menghitung outlier gucheng_df
for column in columns:
    q25, q75 = np.percentile(gucheng_df[column], 25), np.percentile(gucheng_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in gucheng_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'gucheng_{column}'] = len(outliers)  # Simpan jumlah outlier

#Menghitung outlier huairou_df
for column in columns:
    q25, q75 = np.percentile(huairou_df[column], 25), np.percentile(huairou_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in huairou_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'huairou_{column}'] = len(outliers)  # Simpan jumlah outlier\

#Menghitung outlier nongzhanguan_df
for column in columns:
    q25, q75 = np.percentile(nongzhanguan_df[column], 25), np.percentile(nongzhanguan_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in nongzhanguan_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'nongzhanguan_{column}'] = len(outliers)  # Simpan jumlah outlier

#Menghitung outlier shunyi_df
for column in columns:
    q25, q75 = np.percentile(shunyi_df[column], 25), np.percentile(shunyi_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in shunyi_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'shunyi_{column}'] = len(outliers)  # Simpan jumlah outlier

#Menghitung outlier tiantan_df
for column in columns:
    q25, q75 = np.percentile(tiantan_df[column], 25), np.percentile(tiantan_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in tiantan_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'tiantan_{column}'] = len(outliers)  # Simpan jumlah outlier

#Menghitung outlier wanliu_df
for column in columns:
    q25, q75 = np.percentile(wanliu_df[column], 25), np.percentile(wanliu_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in wanliu_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'wanliu_{column}'] = len(outliers)  # Simpan jumlah outlier\

#Menghitung outlier wanshouxigong_df
for column in columns:
    q25, q75 = np.percentile(wanshouxigong_df[column], 25), np.percentile(wanshouxigong_df[column], 75)
    iqr = q75 - q25
    cut_off = iqr * 1.5
    minimum, maximum = q25 - cut_off, q75 + cut_off

    outliers = [x for x in wanshouxigong_df[column] if x < minimum or x > maximum]  # Identify outliers
    jumlah_outliers[f'wanshouxigong_{column}'] = len(outliers)  # Simpan jumlah outlier

for column, count in jumlah_outliers.items():
    print(f"Jumlah Outlier di column {column}: {count}")
```

**Mengatasi masalah Inaccurate Value sekaligus mengatasi Outlier**

Menggunakan metode imputasi pada nilai maksimumnya (karena itu yang bermasalah)
pada kolom ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'RAIN', 'WSPM'] menggunakan imputasi mean


```python
# Daftar data station dan kolom yang ingin diproses
setiap_data = [aotizhongxin_df, changping_df, dingling_df, dongsi_df, guanyuan_df, gucheng_df,
              huairou_df, nongzhanguan_df, shunyi_df, tiantan_df, wanliu_df, wanshouxigong_df]
columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'RAIN', 'WSPM']

# Mengganti nilai di setiap data station
for df in setiap_data:
    for column in columns:
        mean_value = df[column].mean()
        std_value = df[column].std()
        lower_bound = mean_value - std_value
        upper_bound = mean_value + std_value

        df.loc[df[column] > upper_bound, column] = mean_value
        df.loc[df[column] < lower_bound, column] = mean_value
```

**Mengatasi Masalah Outlier**

Mengatasi outlier menggunakan metode **mean imputation** pada kolom *'PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'RAIN', dan 'WSPM'*

```python
#Definisikan fungsi untuk menangani outlier dan imputasi
def handle_outliers(df, columns):
    #Hitung kuartil dan IQR
    Q1 = df[columns].quantile(0.25)
    Q3 = df[columns].quantile(0.75)
    IQR = Q3 - Q1

    #Hitung batas maksimum dan minimum
    maximum = Q3 + (1.5 * IQR)
    minimum = Q1 - (1.5 * IQR)

    #Kondisi di bawah minimum atau di atas maksimum
    kondisi_lower_than = df[columns] < minimum
    kondisi_more_than = df[columns] > maximum

    #Hitung nilai mean untuk imputasi
    mean_values = df[columns].mean()

    #Ganti nilai outlier dengan nilai mean
    for col in mean_values.index:
        df[col] = df[col].mask(kondisi_lower_than[col] | kondisi_more_than[col], mean_values[col])

for df in setiap_data:
    handle_outliers(df, columns)
```

**Kapan waktu yang paling tepat untuk beraktivitas di luar ruangan dengan meminimalkan paparan polusi udara?**

Explorasi data kualitas udara disetiap station khususnya kolom 'year', 'month', 'day', 'hour' untuk dijadikan waktu acuan. Dan meng-explore kolom 'PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM' pada setiap data kualitas udara untuk menjadi acuan kualitas udara baik atau tidak

Mengurutkan setiap bulan pada setiap tahun berdasarkan nilai kualitas udara terbaik di setiap station. Memanfaatkan metode `groupby()` untuk menggabungkan rata-rata dari setiap kolom yang terlibat menjadi kolom *avg_bulanan* melakukan sorting menggunakan `sort_values()` dan juga melakukan filtering untuk menampilkan kolom-kolom yang diinginkan menggunakan metode `filter()`

```python
def calculate_monthly_avg(df, station_name):
    #Menggabungkan 'year', 'month', 'day', 'hour' menjadi kolom 'datetime'
    df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])

    #Menambahkan kolom 'tahun_bulan' untuk group by
    df['tahun_bulan'] = df['datetime'].dt.to_period('M')

    #Menambahkan kolom 'station' berdasarkan nama stasiun
    df['station'] = station_name

    #Menghitung rata-rata per bulan
    avg_bulanan = df.groupby(['tahun_bulan', 'station']).agg({
        'PM2.5': 'mean',
        'PM10': 'mean',
        'SO2': 'mean',
        'NO2': 'mean',
        'CO': 'mean',
        'O3': 'mean',
        'TEMP': 'mean',
        'PRES': 'mean',
        'DEWP': 'mean',
        'RAIN': 'mean',
        'WSPM': 'mean'
    }).reset_index()

    #Mengurutkan hasil
    sorted_df = avg_bulanan.sort_values(by=['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM'], ascending=True)

    return sorted_df.head(20)
```

Disini saya membuat fungsi untuk menggabungkan seluruh rata-rata dari setiap polutan dan kondisi alam menjadi satu kesatuan menggunakan metode `groupby()`.

```python
#Fungsi untuk menggabungkan datetime, menghitung mean total
def five_year_avg(df):
    #Menghitung rata-rata per station
    fiveyears_avg = df.groupby('station').agg({
        'PM2.5': 'mean',
        'PM10': 'mean',
        'SO2': 'mean',
        'NO2': 'mean',
        'CO': 'mean',
        'O3': 'mean',
        'TEMP': 'mean',
        'PRES': 'mean',
        'DEWP': 'mean',
        'RAIN': 'mean',
        'WSPM': 'mean'
    }).reset_index()

    return fiveyears_avg
```

Disini saya menggabungkan seluruh data menjadi satu tabel, menggunakan metode `concat()` dan menampilkannya

```python
#Menggabungkan rata-rata dari semua stasiun
all_station_avg = pd.concat([
    five_year_avg(aotizhongxin_df),
    five_year_avg(changping_df),
    five_year_avg(dingling_df),
    five_year_avg(dongsi_df),
    five_year_avg(guanyuan_df),
    five_year_avg(gucheng_df),
    five_year_avg(huairou_df),
    five_year_avg(nongzhanguan_df),
    five_year_avg(shunyi_df),
    five_year_avg(tiantan_df),
    five_year_avg(wanliu_df),
    five_year_avg(wanshouxigong_df)
], ignore_index=True)

#Menghitung rata-rata polutan untuk setiap stasiun
all_station_avg['rank_polutan'] = all_station_avg[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']].mean(axis=1)

#Mengurutkan data berdasarkan 'rank_polutan' dari yang tertinggi ke terendah
all_station_sorted = all_station_avg.sort_values(by='rank_polutan', ascending=False)

#Memfilter data untuk hanya menampilkan kolom yang diinginkan
all_station_avg_filtered = all_station_sorted.filter(items=['station', 'PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM'])

print(all_station_avg_filtered)
```
